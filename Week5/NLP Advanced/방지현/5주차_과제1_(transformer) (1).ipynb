**1. Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: ì£¼ì–´ì§„ ì…ë ¥ì—ì„œ ê° ìš”ì†Œê°€ ë‹¤ë¥¸ ìš”ì†Œì™€ì˜ ê´€ê³„ë¥¼ í•™ìŠµí•˜ëŠ” ë©”ì»¤ë‹ˆì¦˜ì„. ë¬¸ì¥ ë‚´ì—ì„œ ê° ë‹¨ì–´ê°€ ë¬¸ì¥ì˜ ë‹¤ë¥¸ ë‹¨ì–´ì™€ ì–¼ë§ˆë‚˜ ì—°ê´€ë˜ì–´ ìˆëŠ”ì§€ë¥¼ ê³„ì‚°. Query, Key, Value ë²¡í„°ë¥¼ êµ¬ì„±ë˜ë©°, ì´ ë²¡í„°ë“¤ ê°„ì˜ ë‚´ì ì„ í†µí•´ ê° ìš”ì†Œê°€ ë‹¤ë¥¸ ìš”ì†Œì— ì–¼ë§ˆë‚˜ ì§‘ì¤‘í• ì§€ ê°€ì¤‘ì¹˜ë¥¼ ë¶€ì—¬í•¨

```python
import numpy as np

# ê°„ë‹¨í•œ ë¬¸ì¥: ['ë‚˜ëŠ”', 'ì‚¬ê³¼ë¥¼', 'ë¨¹ì—ˆë‹¤']
words = ['ë‚˜ëŠ”', 'ì‚¬ê³¼ë¥¼', 'ë¨¹ì—ˆë‹¤']
word_vectors = {
    'ë‚˜ëŠ”': np.array([1, 0, 0]),  #'ë‚˜ëŠ”'ì— ëŒ€ì‘í•˜ëŠ”ë²¡í„°
    'ì‚¬ê³¼ë¥¼': np.array([0, 1, 0]), #'ì‚¬ê³¼ë¥¼' "
    'ë¨¹ì—ˆë‹¤': np.array([0, 0, 1]) #'ë¨¹ì—ˆë‹¤' "
}

# ìœ ì‚¬ë„ ê³„ì‚° í•¨ìˆ˜ (self-attention)
def self_attention(query, key):
    return np.dot(query, key) #Query ë²¡í„°ì™€ Key ë²¡í„° ê°„ì˜ ë‚´ì  ê³„ì‚°

attention_matrix = np.zeros((len(words), len(words)))
for i in range(len(words)):  #i: Query ë‹¨ì–´
    for j in range(len(words)): #j: Key ë‹¨ì–´
        attention_matrix[i][j] = self_attention(word_vectors[words[i]], word_vectors[words[j]])
         #ië²ˆì§¸ ë‹¨ì–´(Query)ì™€ jë²ˆì§¸ ë‹¨ì–´(Key) ê°„ì˜ self-attentionrê³„ì‚°

print("Self-Attention Matrix:")
print(attention_matrix)
```

**2. Multi-Head Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Multi-Head Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: Self-Attentionì˜ í™•ì¥ëœ í˜•íƒœë¡œ, ì—¬ëŸ¬ ê°œì˜ ì„œë¡œ ë‹¤ë¥¸ Attention Headë¥¼ ì‚¬ìš©í•˜ì—¬ ë” ë‹¤ì–‘í•œ ê´€ê³„ì™€ ì •ë³´ë¥¼ í•™ìŠµí•  ìˆ˜ ìˆë„ë¡ ì„¤ê³„ë¨. Self-Attentionì—ì„œëŠ” ë‹¨ì¼ Attentionë§Œ ê³„ì‚°í•˜ì§€ë§Œ, Multi-Head Attentionì—ì„œëŠ” ì—¬ëŸ¬ê°œì˜ Attention Headë¥¼ ë³‘ë ¬ë¡œ ì‚¬ìš©í•´ ë‹¤ì–‘í•œ ì‹œê°ì—ì„œ ì •ë³´ë¥¼ ì²˜ë¦¬í•¨

```python
# ì—¬ëŸ¬ ê°œì˜ attention heads
def multi_head_self_attention(query, key, heads=3):
    return [np.dot(query, key) for _ in range(heads)]
    #ê° Headì—ì„œ Queryì™€ Keyê°„ì˜ ë‚´ì ì„ ìˆ˜í–‰í•˜ì—¬ Attention ê°’ì„ ê³„ì‚°í•¨

# multi_head_attention_matrixëŠ” ê° ë‹¨ì–´ ê°„ì˜ attentionì„ ì—¬ëŸ¬ headë¡œ í‘œí˜„í•œ í–‰ë ¬
#3ì°¨ì› ë°°ì—´ë¡œ êµ¬ì„±:(ë‹¨ì–´ìˆ˜, ë‹¨ì–´ìˆ˜, headìˆ˜)
multi_head_attention_matrix = np.zeros((len(words), len(words), 3))
for i in range(len(words)):
    for j in range(len(words)):
        multi_head_attention_matrix[i][j] = multi_head_self_attention(word_vectors[words[i]], word_vectors[words[j]])

print("\nMulti-Head Self-Attention Matrix:")
print(multi_head_attention_matrix)
```

**3. Masked Multi-Head Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Masked Multi-Head Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: ê¸°ë³¸ì ì¸ Multi-Head Self Attentionì—ì„œ ë¯¸ë˜ì˜ ì •ë³´ë¥¼ ë§ˆìŠ¤í‚¹í•˜ì—¬ ë³´ì§€ ì•Šë„ë¡ ì„¤ê³„ëœ ë©”ì»¤ë‹ˆì¦˜ì„. ë§ˆìŠ¤í¬ë¥¼ ì‚¬ìš©í•˜ì—¬ íŠ¹ì • ìœ„ì¹˜ì˜ Attentionì„ ì œí•œí•˜ê³ , í˜„ì¬ ì‹œì ì—ì„œ ê³¼ê±° ë‹¨ì–´ì—ë§Œ ì§‘ì¤‘í•˜ë„ë¡ í•˜ëŠ” ë°©ì‹ì„

```python
# ë§ˆìŠ¤í¬ ì¶”ê°€: í˜„ì¬ ë‹¨ì–´ ì´í›„ì˜ ë‹¨ì–´ëŠ” ê³„ì‚°í•˜ì§€ ì•ŠìŒ
def masked_attention(query, key, mask):
    return np.dot(query, key) * mask
    # Queryì™€ Keyê°„ì˜ ë‚´ì ì„ ê³„ì‚°í•˜ê³  ë§ˆìŠ¤í¬ ê°’ì„ ê³±í•¨(ë¯¸ë˜ ë‹¨ì–´ëŠ” 0ìœ¼ë¡œ ì²˜ë¦¬), maskê°€ 0ì¸ ê²½ìš° í•´ë‹¹ attentionê°’ì„ ë¬´ì‹œí•¨

mask = np.array([1, 1, 0])  # ì²« ë²ˆì§¸, ë‘ ë²ˆì§¸ëŠ” ë³´ì§€ë§Œ, ì„¸ ë²ˆì§¸ëŠ” ë‹¨ì–´ëŠ” ë³´ì§€ì•ŠìŒ
masked_attention_matrix = np.zeros((len(words), len(words)))
for i in range(len(words)):
    for j in range(len(words)):
        masked_attention_matrix[i][j] = masked_attention(word_vectors[words[i]], word_vectors[words[j]], mask[j])

print("\nMasked Self-Attention Matrix:")
print(masked_attention_matrix)
```

**4. Cross AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Cross Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: ë‘ ê°œì˜ ì„œë¡œ ë‹¤ë¥¸ ì‹œí€€ìŠ¤ ê°„ì— ì´ë£¨ì–´ì§€ëŠ” Attention ë©”ì»¤ë‹ˆì¦˜ì„. ì´ëŠ” ì£¼ë¡œ ì§ˆë¬¸ê³¼ ì‘ë‹µ, ë˜ëŠ” ì…ë ¥ê³¼ ì¶œë ¥ ê°™ì€ ìƒí˜¸ì‘ìš©ì´ í•„ìš”í•œ ìƒí™©ì—ì„œ ì‚¬ìš©ë¨. self Attentionì´ ë‹¨ì¼ ì‹œí€€ìŠ¤ ë‚´ì—ì„œ ë‹¨ì–´ë“¤ ê°„ì˜ ìƒí˜¸ì‘ìš©ì„ ê³„ì‚°í•˜ëŠ” ë°˜ë©´, Cross Attentionì€ ë‘ ê°œì˜ ì„œë¡œ ë‹¤ë¥¸ ì‹œí€€ìŠ¤ì—ì„œ ë‹¨ì–´ë“¤ ê°„ì˜ ì—°ê´€ì„±ì„ ê³„ì‚°í•¨

```python
# ì…ë ¥ ë¬¸ì¥ê³¼ ì‘ë‹µ ë¬¸ì¥
question_words = ['ë„ˆëŠ”', 'ì‚¬ê³¼ë¥¼']
answer_words = ['ë‚˜ëŠ”', 'ë¨¹ì—ˆë‹¤']
question_vectors = {
    'ë„ˆëŠ”': np.array([1, 0]),
    'ì‚¬ê³¼ë¥¼': np.array([0, 1])
}
answer_vectors = {
    'ë‚˜ëŠ”': np.array([1, 0]),
    'ë¨¹ì—ˆë‹¤': np.array([0, 1])
}

# Cross-Attention
cross_attention_matrix = np.zeros((len(question_words), len(answer_words)))
for i in range(len(question_words)):
    for j in range(len(answer_words)):
        cross_attention_matrix[i][j] = np.dot(question_vectors[question_words[i]], answer_vectors[answer_words[j]])

print("\nCross-Attention Matrix:")
print(cross_attention_matrix)
```
