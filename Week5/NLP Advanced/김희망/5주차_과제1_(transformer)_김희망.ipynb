**1. Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: token vectorë¼ë¦¬ì˜ ë‚´ì , ì¦‰ ìœ ì‚¬ë„ë¥¼ ê³„ì‚°í•˜ì—¬ ê° tokenë“¤ì´ ì–¼ë§ˆë‚˜ ë¹„ìŠ·í•œì§€ ìˆ«ìë¡œ ë‚˜íƒ€ë‚´ëŠ” ë©”ì»¤ë‹ˆì¦˜ì´ë‹¤.

```python
import numpy as np

# ê°„ë‹¨í•œ ë¬¸ì¥: ['ë‚˜ëŠ”', 'ì‚¬ê³¼ë¥¼', 'ë¨¹ì—ˆë‹¤']
words = ['ë‚˜ëŠ”', 'ì‚¬ê³¼ë¥¼', 'ë¨¹ì—ˆë‹¤'] # token list
word_vectors = {
    'ë‚˜ëŠ”': np.array([1, 0, 0]),
    'ì‚¬ê³¼ë¥¼': np.array([0, 1, 0]),
    'ë¨¹ì—ˆë‹¤': np.array([0, 0, 1])
}

# ìœ ì‚¬ë„ ê³„ì‚° í•¨ìˆ˜ (self-attention)
def self_attention(query, key):
    return np.dot(query, key) # ë²¡í„°ì˜ ë‚´ì  ê°’ ë°˜í™˜

attention_matrix = np.zeros((len(words), len(words)))
for i in range(len(words)):
    for j in range(len(words)):
        attention_matrix[i][j] = self_attention(word_vectors[words[i]], word_vectors[words[j]]) # ë‚´ì ì„ ì´ìš©í•˜ì—¬ ë²¡í„°ê°„ì˜ ìœ ì‚¬ë„ë¥¼ ê³„ì‚°,

print("Self-Attention Matrix:")
print(attention_matrix)
```

**2. Multi-Head Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Multi-Head Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: ì—¬ëŸ¬ ë¬¸ë§¥ì„ ê³ ë ¤í•˜ì—¬ ì—¬ëŸ¬ë²ˆ self attentionì„ ìˆ˜í–‰í•˜ì—¬ ë¬¸ë§¥ì„ ë”ìš± íš¨ê³¼ì ìœ¼ë¡œ íŒŒì•…í•˜ë„ë¡ í•˜ëŠ” ë©”ì»¤ë‹ˆì¦˜

```python
# ì—¬ëŸ¬ ê°œì˜ attention heads
def multi_head_self_attention(query, key, heads=3):
    return [np.dot(query, key) for _ in range(heads)] # self-attentionì„ 3ë²ˆ ë°˜ë³µ

multi_head_attention_matrix = np.zeros((len(words), len(words), 3))
for i in range(len(words)):
    for j in range(len(words)):
        multi_head_attention_matrix[i][j] = multi_head_self_attention(word_vectors[words[i]], word_vectors[words[j]]) #multi-head self attentionì„ ì´ìš©í•˜ì—¬ ì¿¼ë¦¬ì™€ í‚¤ì˜ ë‚´ì ê°’ì„ ê³„ì‚°

print("\nMulti-Head Self-Attention Matrix:")
print(multi_head_attention_matrix)
```

**3. Masked Multi-Head Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Masked Multi-Head Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: transformer í›ˆë ¨ ê³¼ì •ì—ì„œ ì •ë‹µì´ decoderì˜ inputê°’ìœ¼ë¡œ ë“¤ì–´ê°€ë¯€ë¡œ ì •ë‹µì„ ì°¸ê³ í•˜ì—¬ ì •ë‹µì„ returní•˜ì§€ ì•Šë„ë¡ íŠ¹ì • ë‹¨ì–´ë¥¼ ê³¼ì •ì—ì„œ ì œì™¸ì‹œí‚´

```python
# ë§ˆìŠ¤í¬ ì¶”ê°€: í˜„ì¬ ë‹¨ì–´ ì´í›„ì˜ ë‹¨ì–´ëŠ” ê³„ì‚°í•˜ì§€ ì•ŠìŒ
def masked_attention(query, key, mask):
    return np.dot(query, key) * mask # ë§Œì•½ 3ë²ˆì§¸ keyì™€ ë‚´ì í•˜ëŠ” ê²½ìš° 0ì„ ê³±í•˜ì—¬ ìµœì¢… ê°’ì„ 0ìœ¼ë¡œ ë§Œë“¦

mask = np.array([1, 1, 0])  # ì²« ë²ˆì§¸, ë‘ ë²ˆì§¸ëŠ” ë³´ì§€ë§Œ, ì„¸ ë²ˆì§¸ëŠ” ì œì™¸í•´ì„œ ë´„ (ì´ ë¹ˆì¹¸ ê¼­ ì±„ì›Œì£¼ì„¸ìš” :) )
masked_attention_matrix = np.zeros((len(words), len(words)))
for i in range(len(words)):
    for j in range(len(words)):
        masked_attention_matrix[i][j] = masked_attention(word_vectors[words[i]], word_vectors[words[j]], mask[j])

print("\nMasked Self-Attention Matrix:")
print(masked_attention_matrix)
```

**4. Cross AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Cross Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: self-attentionì—ì„œ ê°™ì€ ë‹¨ì–´ë¼ë¦¬ì˜ ê³„ì‚°ì´ ì¼ì–´ë‚˜ëŠ” ê²ƒì„ ë°©ì§€í•˜ê³  íŠ¹ì • ë¬¸ì¥ê³¼ ë‹¤ë¥¸ ë¬¸ì¥ì˜ ê´€ê³„ë¥¼ íŒŒì•…í•˜ê¸° ìœ„í•´ crossí•˜ì—¬ attentionì„ ì§„í–‰

```python
# ì…ë ¥ ë¬¸ì¥ê³¼ ì‘ë‹µ ë¬¸ì¥
question_words = ['ë„ˆëŠ”', 'ì‚¬ê³¼ë¥¼'] # ë¬¸ì¥1
answer_words = ['ë‚˜ëŠ”', 'ë¨¹ì—ˆë‹¤'] # ë¬¸ì¥2
question_vectors = {
    'ë„ˆëŠ”': np.array([1, 0]),
    'ì‚¬ê³¼ë¥¼': np.array([0, 1])
}
answer_vectors = {
    'ë‚˜ëŠ”': np.array([1, 0]),
    'ë¨¹ì—ˆë‹¤': np.array([0, 1])
}

# Cross-Attention
cross_attention_matrix = np.zeros((len(question_words), len(answer_words)))
for i in range(len(question_words)):
    for j in range(len(answer_words)):
        cross_attention_matrix[i][j] = np.dot(question_vectors[question_words[i]], answer_vectors[answer_words[j]]) # ë¬¸ì¥1ì˜ ë‹¨ì–´ë“¤ê³¼ ë¬¸ì¥ 2ì˜ ë‹¨ì–´ë“¤ì„ ì°¨ë¡€ë¡œ ë‚´ì í•˜ì—¬ ìœ ì‚¬ë„ ê³„ì‚°

print("\nCross-Attention Matrix:")
print(cross_attention_matrix)
```
