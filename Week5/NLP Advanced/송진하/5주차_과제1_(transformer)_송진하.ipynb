**1. Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: Self Attentionì€ ì…ë ¥ ë¬¸ì¥ì—ì„œ ê° ë‹¨ì–´ê°€ ë¬¸ì¥ ë‚´ ë‹¤ë¥¸ ë‹¨ì–´ë“¤ê³¼ì˜ ê´€ê³„ë¥¼ ì–¼ë§ˆë‚˜ ì¤‘ìš”í•˜ê²Œ ì—¬ê²¨ì•¼ í•˜ëŠ”ì§€ë¥¼ í•™ìŠµí•˜ëŠ” ë°©ë²•ì´ë‹¤.

```python
import numpy as np

# ê°„ë‹¨í•œ ë¬¸ì¥: ['ë‚˜ëŠ”', 'ì‚¬ê³¼ë¥¼', 'ë¨¹ì—ˆë‹¤']
words = ['ë‚˜ëŠ”', 'ì‚¬ê³¼ë¥¼', 'ë¨¹ì—ˆë‹¤']
word_vectors = {
    'ë‚˜ëŠ”': np.array([1, 0, 0]),
    'ì‚¬ê³¼ë¥¼': np.array([0, 1, 0]), # ê° ë‹¨ì–´ì— ëŒ€í•œ ì„ë² ë”©
    'ë¨¹ì—ˆë‹¤': np.array([0, 0, 1])
}

# ìœ ì‚¬ë„ ê³„ì‚° í•¨ìˆ˜ (self-attention)
# queryì™€ key ë²¡í„° ê°„ì˜ ë‚´ì (dot product)ì„ í†µí•´ ìœ ì‚¬ë„ë¥¼ ê³„ì‚°
def self_attention(query, key):
    return np.dot(query, key)

attention_matrix = np.zeros((len(words), len(words)))
# ë¬¸ì¥ ë‚´ì˜ ëª¨ë“  ë‹¨ì–´ ìŒì— ëŒ€í•´ self-attention scoreë¥¼ ê³„ì‚°
for i in range(len(words)):
    for j in range(len(words)):
      # ê° ë‹¨ì–´ ìŒ (queryì™€ key) ê°„ì˜ attention scoreë¥¼ ê³„ì‚°
        attention_matrix[i][j] = self_attention(word_vectors[words[i]], word_vectors[words[j]])

# ìµœì¢… self-attention í–‰ë ¬ ì¶œë ¥
print("Self-Attention Matrix:")
print(attention_matrix)
```

**2. Multi-Head Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Multi-Head Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: Self Attentionì„ ì—¬ëŸ¬ ê°œì˜ headë¡œ ë‚˜ëˆ„ì–´ ë³‘ë ¬ë¡œ ìˆ˜í–‰í•˜ëŠ” ë°©ì‹ì´ë‹¤. ê° headëŠ” ì…ë ¥ì„ ë‹¤ë¥¸ ë°©ì‹ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì„œë¡œ ë‹¤ë¥¸ ë¶€ë¶„ì˜ ì •ë³´ë¥¼ í•™ìŠµí•œë‹¤.

```python
# ì—¬ëŸ¬ ê°œì˜ attention heads
def multi_head_self_attention(query, key, heads=3):
  # queryì™€ keyì˜ ë‚´ì (dot product)ì„ ì—¬ëŸ¬ ë²ˆ(heads ê°œìˆ˜ë§Œí¼) ê³„ì‚°í•˜ì—¬ ê° headì— ëŒ€í•´ ìœ ì‚¬ë„ë¥¼ êµ¬í•¨
    return [np.dot(query, key) for _ in range(heads)] # ê° headì—ì„œ ë™ì¼í•œ query, keyì— ëŒ€í•œ attentionì„ ê³„ì‚°


multi_head_attention_matrix = np.zeros((len(words), len(words), 3))
# ë¬¸ì¥ ë‚´ì˜ ëª¨ë“  ë‹¨ì–´ ìŒì— ëŒ€í•´ multi-head self-attention scoreë¥¼ ê³„ì‚°
for i in range(len(words)):
    for j in range(len(words)):
        multi_head_attention_matrix[i][j] = multi_head_self_attention(word_vectors[words[i]], word_vectors[words[j]])
# ìµœì¢… multi-head self-attention í–‰ë ¬ ì¶œë ¥
print("\nMulti-Head Self-Attention Matrix:")
print(multi_head_attention_matrix)
```

**3. Masked Multi-Head Self AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Masked Multi-Head Self Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…: Self Attentionì—ì„œ ì¼ë¶€ ë‹¨ì–´ì— ëŒ€í•œ ì •ë³´ë¥¼ ë³´ì§€ ëª»í•˜ë„ë¡ maskingì„ ì ìš©í•˜ëŠ” ë©”ì»¤ë‹ˆì¦˜ì´ë‹¤. ì´ ë°©ë²•ì€ íŠ¹íˆ ì–¸ì–´ ëª¨ë¸ì—ì„œ ë¯¸ë˜ì˜ ë‹¨ì–´ë¥¼ ì˜ˆì¸¡í•  ë•Œ ìœ ìš©í•˜ë‹¤.

```python
# ë§ˆìŠ¤í¬ ì¶”ê°€: í˜„ì¬ ë‹¨ì–´ ì´í›„ì˜ ë‹¨ì–´ëŠ” ê³„ì‚°í•˜ì§€ ì•ŠìŒ
def masked_attention(query, key, mask):
  # queryì™€ key ê°„ì˜ ë‚´ì  ê³„ì‚° í›„ mask ê°’ì„ ê³±í•˜ì—¬ attention ê³„ì‚°
    return np.dot(query, key) * mask

mask = np.array([1, 1, 0])  # ì²« ë²ˆì§¸, ë‘ ë²ˆì§¸ëŠ” ë³´ì§€ë§Œ, ì„¸ ë²ˆì§¸ëŠ” ë³´ì§€ ì•ŠìŒ (ì´ ë¹ˆì¹¸ ê¼­ ì±„ì›Œì£¼ì„¸ìš” :) )
masked_attention_matrix = np.zeros((len(words), len(words))) # Masked Attention í–‰ë ¬ ìƒì„±
for i in range(len(words)): # ë¬¸ì¥ ë‚´ì˜ ëª¨ë“  ë‹¨ì–´ ìŒì— ëŒ€í•´ masked self-attention scoreë¥¼ ê³„ì‚°
    for j in range(len(words)):
      # mask[j]ë¥¼ ì ìš©í•˜ì—¬ ê° ë‹¨ì–´ ìŒì˜ attentionì„ ê³„ì‚°
        masked_attention_matrix[i][j] = masked_attention(word_vectors[words[i]], word_vectors[words[j]], mask[j])

# ìµœì¢… masked self-attention í–‰ë ¬ ì¶œë ¥
print("\nMasked Self-Attention Matrix:")
print(masked_attention_matrix)
```

**4. Cross AttentionğŸ“Œ**

ì•„ë˜ ì½”ë“œë¥¼ ìˆ˜í–‰í•´ë³´ê³ , Cross Attentionì€ ì–´ë–¤ ë©”ì»¤ë‹ˆì¦˜ì¸ì§€ ì„¤ëª…í•˜ê³ ,

 ê·¸ ì„¤ëª…ì— ë§ê²Œ ê° ì½”ë“œì— ì§ì ‘ ì£¼ì„ì„ ë‹¬ì•„ë´…ì‹œë‹¤.

 âœ…ì„¤ëª…:

```python
# ì…ë ¥ ë¬¸ì¥ê³¼ ì‘ë‹µ ë¬¸ì¥
question_words = ['ë„ˆëŠ”', 'ì‚¬ê³¼ë¥¼'] # ì§ˆë¬¸ ë‹¨ì–´ë“¤
answer_words = ['ë‚˜ëŠ”', 'ë¨¹ì—ˆë‹¤'] # ì‘ë‹µ ë‹¨ì–´ë“¤
# ê° ì§ˆë¬¸ ë‹¨ì–´ì— ëŒ€í•œ ë²¡í„°
question_vectors = {
    'ë„ˆëŠ”': np.array([1, 0]),
    'ì‚¬ê³¼ë¥¼': np.array([0, 1])
}
# ê° ì‘ë‹µ ë‹¨ì–´ì— ëŒ€í•œ ë²¡í„°
answer_vectors = {
    'ë‚˜ëŠ”': np.array([1, 0]),
    'ë¨¹ì—ˆë‹¤': np.array([0, 1])
}

# Cross-Attention
cross_attention_matrix = np.zeros((len(question_words), len(answer_words)))
# ì§ˆë¬¸ ë‹¨ì–´ì™€ ì‘ë‹µ ë‹¨ì–´ ê°„ì˜ Attention ê³„ì‚°
for i in range(len(question_words)):
    for j in range(len(answer_words)):
        cross_attention_matrix[i][j] = np.dot(question_vectors[question_words[i]], answer_vectors[answer_words[j]]) # ê° ì§ˆë¬¸ ë‹¨ì–´ì™€ ì‘ë‹µ ë‹¨ì–´ì˜ ë²¡í„° ê°„ ë‚´ì ì„ ê³„ì‚°í•˜ì—¬ Attention scoreë¥¼ êµ¬í•¨
# ìµœì¢… Cross-Attention í–‰ë ¬ ì¶œë ¥
print("\nCross-Attention Matrix:")
print(cross_attention_matrix)
```
