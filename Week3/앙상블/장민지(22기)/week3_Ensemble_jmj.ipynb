## Q1. High varianceÏùò Ï†ïÏùò Î∞è Î¨∏Ï†úÏ†êÏóê ÎåÄÌï¥ ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî.

üêª‚Äç‚ùÑÔ∏è High VarianceÎäî Î®∏Ïã†Îü¨Îãù Î™®Îç∏Ïù¥ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Ïóê Í≥ºÎèÑÌïòÍ≤å ÎßûÏ∂∞Ï†∏(overfitting) ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ ÏùºÎ∞òÌôîÌïòÏßÄ Î™ªÌïòÎäî ÏÉÅÌÉúÎ•º ÏùòÎØ∏Ìï©ÎãàÎã§. üêª‚Äç‚ùÑÔ∏è High BiasÎäî Î®∏Ïã†Îü¨Îãù Î™®Îç∏Ïù¥ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Ïùò Ï§ëÏöîÌïú Ìå®ÌÑ¥ÏùÑ Ï∂©Î∂ÑÌûà ÌïôÏäµÌïòÏßÄ Î™ªÌï¥, ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ÏôÄ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î™®ÎëêÏóêÏÑú Î∂ÄÏ†ïÌôïÌïú ÏòàÏ∏°ÏùÑ ÎÇ¥Îäî ÏÉÅÌÉúÎ•º ÏùòÎØ∏Ìï©ÎãàÎã§. Ïù¥Îäî Î™®Îç∏Ïù¥ ÌäπÏ†ï ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Ïóê Í≥ºÎèÑÌïòÍ≤å ÏùòÏ°¥ÌïòÏßÄ ÏïäÎèÑÎ°ù ÌïòÏó¨ high variance problemÏùÑ Ìï¥Í≤∞Ìï©ÎãàÎã§.



- High bias problem : üêª‚Äç‚ùÑÔ∏è BoostingÏùÄ Ïó¨Îü¨ ÏïΩÌïú ÌïôÏäµÏûêÎ•º ÏàúÏ∞®Ï†ÅÏúºÎ°ú ÌïôÏäµÏãúÌÇ§Î©¥ÏÑú, Ïù¥Ï†Ñ Î™®Îç∏Ïù¥ ÏûòÎ™ª ÏòàÏ∏°Ìïú Îç∞Ïù¥ÌÑ∞Ïóê Îçî ÎßéÏùÄ Í∞ÄÏ§ëÏπòÎ•º Î∂ÄÏó¨ÌïòÏó¨ ÌõÑÏÜç Î™®Îç∏Ïù¥ Ïù¥Î•º Í∞úÏÑ†ÌïòÎèÑÎ°ù Ìï©ÎãàÎã§. Ïù¥Î†áÍ≤å ÌïôÏäµÏûêÎì§ÏùÑ ÏàúÏ∞®Ï†ÅÏúºÎ°ú Í≤∞Ìï©Ìï®ÏúºÎ°úÏç® Ï†êÏ†ê Îçî ÏÑ±Îä•Ïù¥ Í∞úÏÑ†Îêú Î™®Îç∏ÏùÑ ÎßåÎì§Ïñ¥Í∞ëÎãàÎã§. ## EDA & Preprocessing

EDAÎ•º ÌÜµÌï¥ Ïù∏ÏÇ¨Ïù¥Ìä∏Î•º ÏñªÍ≥† Ï†ÑÏ≤òÎ¶¨Î•º ÏßÑÌñâÌï¥Ï£ºÏÑ∏Ïöî.

````python
# ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ Î°úÎìú
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Îç∞Ïù¥ÌÑ∞Ïùò Í∏∞Î≥∏ Ï†ïÎ≥¥ ÌôïÏù∏
print(df.info())

# Í∏∞Ï¥à ÌÜµÍ≥ÑÎüâ ÌôïÏù∏
print(df.describe())

# Í≤∞Ï∏°Í∞í ÌôïÏù∏
print(df.isnull().sum())

# ÌÉÄÍ≤ü Î≥ÄÏàòÏùò Î∂ÑÌè¨ ÌôïÏù∏ (Ïòà: Í≥†Í∞ù Ïù¥ÌÉà Ïó¨Î∂ÄÍ∞Ä TargetÏù∏ Í≤ΩÏö∞)
sns.countplot(df['Attrition_Flag'])
plt.title('Attrition Flag Distribution')
plt.show()

# Ïà´ÏûêÌòï Ïª¨ÎüºÎßå ÏÑ†ÌÉùÌï¥ÏÑú ÏÉÅÍ¥ÄÌñâÎ†¨ Í≥ÑÏÇ∞
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# ÏÉÅÍ¥ÄÌñâÎ†¨ Í≥ÑÏÇ∞
corr_matrix = numeric_df.corr()

# ÏÉÅÍ¥ÄÌñâÎ†¨ ÌûàÌä∏Îßµ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Numeric Features')
plt.show()



# Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ ÌôïÏù∏
for col in df.select_dtypes(include=['object']).columns:
    sns.countplot(df[col])
    plt.title(f'{col} Distribution')
    plt.show()

# ÏàòÏπòÌòï Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ ÌôïÏù∏
df.hist(bins=30, figsize=(15, 10))
plt.show()
````
