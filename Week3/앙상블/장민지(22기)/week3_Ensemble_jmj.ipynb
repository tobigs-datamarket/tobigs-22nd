## Q1. High variance의 정의 및 문제점에 대해 설명해주세요.

🐻‍❄️ High Variance는 머신러닝 모델이 학습 데이터에 과도하게 맞춰져(overfitting) 새로운 데이터에 대해 일반화하지 못하는 상태를 의미합니다. 즉, 모델이 학습 데이터에 지나치게 적합하여 훈련 데이터에서 높은 성능을 보이지만, 테스트 데이터나 새로운 데이터에서는 성능이 급격히 떨어지는 문제를 나타냅니다.

## Q2. High bias의 정의 및 문제점에 대해 설명해주세요.

🐻‍❄️ High Bias는 머신러닝 모델이 학습 데이터의 중요한 패턴을 충분히 학습하지 못해, 훈련 데이터와 테스트 데이터 모두에서 부정확한 예측을 내는 상태를 의미합니다. 이는 underfitting(과소적합) 상태라고도 불립니다. High Bias는 모델이 너무 단순해서 데이터를 잘 설명하지 못하는 경우 발생합니다.

## Q3. High variance 와 high bias를 다룰때 적합한 ensemble 방식과 그 이유를 적어주세요.

- High variance problem : 🐻‍❄️ Bagging은 여러 약한 학습자(모델)를 병렬로 학습시키고, 이들의 예측을 평균화(회귀)하거나 다수결 투표(분류)를 통해 최종 결과를 도출합니다. 각 모델은 학습 데이터를 랜덤하게 샘플링하여 훈련되므로, 학습 데이터의 작은 변화가 각 모델에 미치는 영향을 분산시킵니다. 이는 모델이 특정 학습 데이터에 과도하게 의존하지 않도록 하여 high variance problem을 해결합니다.



- High bias problem : 🐻‍❄️ Boosting은 여러 약한 학습자를 순차적으로 학습시키면서, 이전 모델이 잘못 예측한 데이터에 더 많은 가중치를 부여하여 후속 모델이 이를 개선하도록 합니다. 이렇게 학습자들을 순차적으로 결합함으로써 점점 더 성능이 개선된 모델을 만들어갑니다. 이는 모델이 더 복잡해지도록 만들어 High Bias problem을 해결합니다.

 - Total_Amt_Chng_Q4_Q1 : 거래 금액 변화율 (4분기 대비 1분기)
 - Total_Trans_Amt :  총 거래 금액 (지난 12개월)
 - Total_Trans_Ct : 총 거래 횟수 (지난 12개월)
 - Total_Ct_Chng_Q4_Q1 : 거래 횟수 변화율 (4분기 대비 1분기)
 - Avg_Utilization_Ratio : 평균 카드 이용 비율

[target]
 - Attrition_Flag : 고객 활동 변수 / Existing Customer - 0, Attrited Customer - 1

## EDA & Preprocessing

EDA를 통해 인사이트를 얻고 전처리를 진행해주세요.

````python
# 필수 라이브러리 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# 데이터의 기본 정보 확인
print(df.info())

# 기초 통계량 확인
print(df.describe())

# 결측값 확인
print(df.isnull().sum())

# 타겟 변수의 분포 확인 (예: 고객 이탈 여부가 Target인 경우)
sns.countplot(df['Attrition_Flag'])
plt.title('Attrition Flag Distribution')
plt.show()

# 숫자형 컬럼만 선택해서 상관행렬 계산
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# 상관행렬 계산
corr_matrix = numeric_df.corr()

# 상관행렬 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Numeric Features')
plt.show()



# 범주형 데이터 분포 확인
for col in df.select_dtypes(include=['object']).columns:
    sns.countplot(df[col])
    plt.title(f'{col} Distribution')
    plt.show()

# 수치형 데이터 분포 확인
df.hist(bins=30, figsize=(15, 10))
plt.show()
````
