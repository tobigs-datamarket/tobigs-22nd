##  **2. 데이터 전처리 및 시각화**

- 목표: 목표: 데이터의 품질을 향상시키기 위해 불필요한 데이터를 제거하고, 시계열 데이터의 기본적인 특성(추세 및 변동성)을 시각적으로 탐색합니다.

**지시사항** :
- 거래량(Volume)이 0인 일자는 데이터를 제거하여 품질을 향상시키세요.
- 수정종가(Adj Close) 데이터를 사용하여 50일 이동평균(Rolling Mean)과 이동표준편차(Rolling Standard Deviation)를 계산하세요.
- 원본 데이터와 함께 이동평균과 이동표준편차를 시각화하고, 시각화 결과를 바탕으로 시계열 데이터의 추세와 변동성에 대해 논의하세요.


```python
# 거래량이 0인 일자 제거 및 수정종가 데이터만 사용
data = data[data['Volume'] != 0]

# 이동평균 및 이동표준편차 시각화 함수
def plot_rolling(data, interval):
    rolmean = data.rolling(interval).mean()
    rolstd = data.rolling(interval).std()
    plt.figure(figsize=(10, 6))
    plt.plot(data, color='blue', label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean {}'.format(interval))
    plt.plot(rolstd, color='black', label='Rolling Std {}'.format(interval))
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()

# 50일치 이동평균 및 이동표준편차 계산
plot_rolling(data, interval=50)
```

## **3. 시계열 데이터의 정상성 확인 - ADF 테스트**

### **목표**
시계열 데이터의 정상성 여부를 확인합니다. 정상성은 예측 모델을 적용하기 위한 필수 조건으로, 데이터를 적절히 변환하여 예측 모델이 더 잘 작동할 수 있도록 합니다.

### **지시사항**:

- ADF(Augmented Dickey-Fuller) 테스트를 사용하여 원본 데이터의 정상성 여부를 확인하세요.
- 1차 차분 데이터를 생성하고, 1차 차분 데이터에 대해 ADF 테스트를 수행하여 정상성 여부를 확인하세요.
- ADF 테스트의 p-value 결과를 바탕으로, 원본 데이터와 1차 차분 데이터의 정상성 여부를 논의하세요.


```python
# ADF 테스트 함수 정의
def adf_test(data):
    result = adfuller(data.values)
    print('ADF Statistics: %f' % result[0])
    print('p-value: %f' % result[1])
    return result[1]

# ADF 테스트 수행 - 원본 데이터
print("ADF Test - 원본 데이터")
p_value_original = adf_test(data['Adj Close'])  # 원본 데이터를 사용하세요.

# 1차 차분 데이터 생성 및 ADF 테스트 수행
diff1 = data['Adj Close'].diff().dropna()
print("\nADF Test - 1차 차분 데이터")
p_value_diff1 = adf_test(diff1)  # 1차 차분 데이터를 사용하세요.
```

- A. 평가 지표 중 'MAPE'가 가장 중요한 지표로 생각됨. 그 이유는 MAPE는 실제 값 대비 예측 값의 오차 비율을 백분율로 나타내므로, 데이터의 스케일에 의존하지 않고 비교할 수 있기 때문임. 특히 주가 예측에서는 주가 수준 자체가 변화하더라도, 오차가 상대적으로 얼마나 큰지 파악할 수 있어 실무에서 많이 사용됨. 또한, MAPE는 예측이 어느 정도의 정확도를 가지고 있는지 직관적으로 해석할 수 있는 장점이 있음.

이 과제는 시계열 분석에 대한 이해를 심화하고 Prophet 모델을 실제로 적용해 보고 싶은 분들을 위한 선택 과제입니다. 강의에서 다루지 않았지만, 시계열 데이터 분석의 다양한 가능성을 탐구하고 싶은 분들은 도전해 보세요! 😊


```python
# 변수 중복으로 인하여 한 코드로 작성함
# 문제 1: 주가 데이터 수집 및 전처리
import yfinance as yf
import pandas as pd

# Yahoo Finance에서 Google 주가 데이터를 수집
ticker = 'GOOGL'
start_date = '2014-06-01'
end_date = '2024-06-01'
data = yf.download(ticker, start=start_date, end=end_date)

# 'Adj Close' 열만 사용하여 Prophet 모델이 요구하는 포맷으로 변환
data = data[['Adj Close']].reset_index()
data.columns = ['ds', 'y']  # Prophet의 요구 형식 (ds: 날짜, y: 값)

# 데이터 확인
print("문제 1 데이터 전처리 결과:")
print(data.head())

# 문제 2: Prophet 모델 학습
from prophet import Prophet

# Prophet 모델 초기화 및 학습
model = Prophet()
model.fit(data)

print("문제 2 Prophet 모델 학습 완료")

# 문제 3: 미래 주가 예측
# 향후 1년 동안의 주가 예측
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# 예측 결과 시각화
print("문제 3 미래 주가 예측 결과:")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())

# 예측 시각화
fig = model.plot(forecast)
fig.show()

# 문제 4: 모델 성능 평가
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# 테스트 데이터는 마지막 10%로 분리
train_size = int(len(data) * 0.9)
train_data = data[:train_size]
test_data = data[train_size:]

# 예측값 가져오기 (학습 데이터 이후의 예측값 사용)
predictions = forecast['yhat'][train_size:]

# 평가 지표 계산
mse = mean_squared_error(test_data['y'], predictions)
mae = mean_absolute_error(test_data['y'], predictions)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((test_data['y'] - predictions) / test_data['y'])) * 100

# 성능 지표 출력
print(f'MSE: {mse}, MAE: {mae}, RMSE: {rmse}, MAPE: {mape}')

#추가 질문:

# 1. Prophet 모델의 예측 정확도를 높이기 위해 추가할 수 있는 요소는 무엇인가요?
# A. 추가로 고려할 수 있는 요소는 외부 변수(메타 데이터)임. 예를 들어, 경제 지표나 기업 실적을 포함하여 예측을 개선할 수 있음.

# 2. Prophet 모델은 비선형적인 패턴과 외부 요인을 잘 반영하는가요? 그 이유를 설명하세요.
# A. Prophet 모델은 주로 계절성과 트렌드 변화를 잘 포착하지만, 비선형적인 외부 요인에는 한계가 있음. 비선형 패턴을 반영하려면 추가적인 외부 요인을 모델에 포함시켜야 함.
```
